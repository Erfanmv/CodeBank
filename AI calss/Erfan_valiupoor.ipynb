```python
import numpy as np
d = 10
m, n = 3, 4
A = np.random.rand (m, d)
B = np.random.rand(n, d)
```

```python
C = np.zeros((m,n))
for i in range(m):
    for j in range(n):
        C[i, j] = np.sum(A[i] * B[j])
C
```

```python
C = np.dot(A, B.T)
C
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("dataexam.csv")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```

جواب سوال دوم دو

```python
num_rows, num_columns = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

جواب سوال دوم سه

```python
average_salary = df['AnnualIncome'].mean()
print("Average salary of clients:", average_salary)
```

جواب سوال دوم چهار

```python
# Count the occurrences of each unique value in the 'TravelInsurance' column
insurance_counts = df['TravelInsurance'].value_counts()

# Extract counts for people who bought and didn't buy travel insurance
bought_insurance = insurance_counts.get(1, 0)  # Assuming 1 represents bought insurance
didnt_buy_insurance = insurance_counts.get(0, 0)  # Assuming 0 represents didn't buy insurance

# Print the counts
print("Number of people who bought travel insurance:", bought_insurance)
print("Number of people who didn't buy travel insurance:", didnt_buy_insurance)

# Conclude based on the counts
if bought_insurance > didnt_buy_insurance:
    print("Most people are insured.")
else:
    print("Most people are not insured.")
```

جواب سوال دوم پنچ

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Normalize the data
normalized_df = df.copy()
normalized_df['Age_normalized'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
normalized_df['FamilyMembers_normalized'] = (df['FamilyMembers'] - df['FamilyMembers'].min()) / (df['FamilyMembers'].max() - df['FamilyMembers'].min())
normalized_df['AnnualIncome_normalized'] = (df['AnnualIncome'] - df['AnnualIncome'].min()) / (df['AnnualIncome'].max() - df['AnnualIncome'].min())

# Map colors based on GraduateOrNot
color_map = {'Yes': 'blue', 'No': 'red'}
normalized_df['color'] = df['GraduateOrNot'].map(color_map)

# Draw the scatter plot with original x and y values
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['FamilyMembers'], s=normalized_df['AnnualIncome_normalized']*1000, c=normalized_df['color'], alpha=0.5)
plt.title('Scatter Plot of Age vs Family Members')
plt.xlabel('Age')
plt.ylabel('Family Members')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Graduated'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Not Graduated')])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Selecting numerical columns to normalize
numerical_cols = ['Age', 'FamilyMembers', 'AnnualIncome']

# Normalizing the data
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Map colors based on GraduateOrNot
color_map = {'Yes': 'blue', 'No': 'red'}
df['color'] = df['GraduateOrNot'].map(color_map)

# Draw the scatter plot with original x and y values
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['FamilyMembers'], s=df['AnnualIncome']*1000, c=df['color'], alpha=0.5)
plt.title('Scatter Plot of Age vs Family Members')
plt.xlabel('Age')
plt.ylabel('Family Members')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Graduated'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Not Graduated')])
plt.show()
```
