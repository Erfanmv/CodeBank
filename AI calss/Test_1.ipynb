```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("dataexam.csv")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```

```python
num_rows, num_columns = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
average_salary = df['AnnualIncome'].mean()
print("Average salary of clients:", average_salary)
```

```python
# Count the occurrences of each unique value in the 'TravelInsurance' column
insurance_counts = df['TravelInsurance'].value_counts()

# Extract counts for people who bought and didn't buy travel insurance
bought_insurance = insurance_counts.get(1, 0)  # Assuming 1 represents bought insurance
didnt_buy_insurance = insurance_counts.get(0, 0)  # Assuming 0 represents didn't buy insurance

# Print the counts
print("Number of people who bought travel insurance:", bought_insurance)
print("Number of people who didn't buy travel insurance:", didnt_buy_insurance)

# Conclude based on the counts
if bought_insurance > didnt_buy_insurance:
    print("Most people are insured.")
else:
    print("Most people are not insured.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mapping for GraduateOrNot to color
color_map = {'Yes': 'blue', 'No': 'red'}

# Swarm plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='FamilyMembers', hue='GraduateOrNot', palette=color_map, size='AnnualIncome', sizes=(50, 500))
plt.title('Swarm Plot of Age vs Family Members with GraduateOrNot and AnnualIncome')
plt.xlabel('Age')
plt.ylabel('Family Members')
plt.legend(title='GraduateOrNot')
plt.show()

```

```python
# Convert 'GraduateOrNot' to categorical for coloring
df['GraduateOrNot'] = df['GraduateOrNot'].astype('category')

# Plotting
sns.set(style="white", color_codes=True)
sns.jointplot(x='Age', y='FamilyMembers', data=df, kind="scatter", hue='GraduateOrNot', palette='Set1', height=7)
plt.xlabel('Age')
plt.ylabel('Family Members')
plt.legend(title='Graduate')
plt.show()
```

```python
# Create the swarm plot
plt.figure(figsize=(10, 6))
ax = sns.swarmplot(data=df, x='Age', y='FamilyMembers', hue='GraduateOrNot', palette={'Yes': 'blue', 'No': 'red'}, size='AnnualIncome', dodge=True)

# Add circle annotations
for _, row in df.iterrows():
    ax.text(row['Age'], row['FamilyMembers'], f"{row['AnnualIncome']}", ha='center', va='center')

# Set labels and title
plt.title('Swarm Plot of Age vs Family Members with GraduateOrNot and AnnualIncome')
plt.xlabel('Age')
plt.ylabel('Family Members')
plt.legend(title='GraduateOrNot')
plt.show()
```

```python
from datetime import datetime

def convert_timestamp_to_datetime(timestamp_ms):
    # Convert milliseconds to seconds
    timestamp_s = timestamp_ms / 1000

    # Convert timestamp to datetime object
    dt_object = datetime.fromtimestamp(timestamp_s)

    return dt_object

# Example usage
timestamp_ms = int(input("Enter the timestamp in milliseconds: "))  # Input timestamp in milliseconds
datetime_obj = convert_timestamp_to_datetime(timestamp_ms)
print("Converted datetime:", datetime_obj)
```

```python
def bytes_to_gb(bytes_value):
    gb_value = bytes_value / (1024**3)  # Convert bytes to gigabytes
    return gb_value

# Input upload and download values in bytes
upload_bytes = int(input("Enter the upload value in bytes: "))
download_bytes = int(input("Enter the download value in bytes: "))

# Convert bytes to gigabytes
upload_gb = bytes_to_gb(upload_bytes)
download_gb = bytes_to_gb(download_bytes)

print("Upload (GB):", upload_gb)
print("Download (GB):", download_gb)
print('tottal traffic: ', upload_gb + download_gb)
```
