```python
import pandas as pd 
import numpy as np
```

```python
movie_df =  pd.read_csv("movies.csv")
rating_df = pd.read_csv("ratings.csv")
print(movie_df.head().to_markdown(numalign="left", stralign="left"))
print(rating_df.head().to_markdown(numalign="left", stralign="left"))
```

```python
movie_df['year'] = movie_df.title.str.extract('(\d\d\d\d)')
movie_df.head()
```

```python
movie_df['title'] = movie_df.title.str.replace('(\(\d\d\d\d\))','',regex=True)
movie_df['title'] = movie_df.title.apply(lambda x : x.strip())
movie_df.head()
```

```python
movierswithgeners_df = movie_df.copy()
for index , row in movie_df.iterrows():
    for genre in row['genres']:
        movierswithgeners_df.at[index,genre]=1
movierswithgeners_df = movierswithgeners_df.fillna(0)
movierswithgeners_df.head()
```

```python
rating_df.head()
```

```python
rating_df = rating_df.drop('timestamp', axis=1,inplace=True)
rating_df
```

```python
userInput = [
    {'title':'Breakfast Club, The','rating':5},
    {'title':'Toy Story','rating':3.5},
    {'title':'Jumanji','rating':2},
    {'title':'Pulp Fiction','rating':5},
    {'title':'Akira','rating':4.5}
]
inputmovies = pd.DataFrame(userInput)
inputmovies
```

```python
inpuId = movie_df[movie_df['title'].isin(inputmovies['title'].tolist())]
inputmovies = pd.merge(inpuId,inputmovies)
inputmovies = inputmovies.drop('genres', axis=1).drop('year',axis=1)
9inputmovies
```

```python
user_movies = movierswithgeners_df[movierswithgeners_df['movieId'].isin(inputmovies['movieId'].tolist())]
user_movies
```
