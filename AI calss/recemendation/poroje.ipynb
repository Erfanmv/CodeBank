```python
moovie['genres']=moovie.genres.str.split('|')
moovie.head()
```

```python
moovierswithgeneres_df=moovie.copy()
for index,row in moovie.iterrows():
    for genre in row['genres']:
        moovierswithgeneres_df.at[index,genre]=1
moovierswithgeneres_df=moovierswithgeneres_df.fillna(0)
moovierswithgeneres_df.head()
```

```python
userinput=[



    {'title': 'Breakfast Club, The', 'rating': 5},
    {'title': 'Toy Story', 'rating': 3.5},
    {'title': 'Jumanji', 'rating': 2},
    {'title': 'Pulp Fiction', 'rating': 5},
    {'title': 'Akira', 'rating': 4.5} 
     


     ]

     


inputmovie=pd.DataFrame(userinput)
inputmovie
```

```python
inputid=moovie[moovie['title'].isin(inputmovie['title'].tolist())]
inputmovie=pd.merge(inputid,inputmovie)
inputmovie=inputmovie.drop('genres',axis=1).drop('year',axis=1)
inputmovie
```

```python
user_movie=moovierswithgeneres_df[moovierswithgeneres_df['movieId'].isin(inputmovie['movieId'].tolist())]
user_movie
```

```python
user_movie = user_movie.reset_index(drop=True)
usergenretable = user_movie.drop('movieId', axis=1).drop('title',axis=1).drop('genres',axis=1).drop('year',axis=1)
usergenretable
```

```python
inputmovie['rating']
```

```python
userprofile = usergenretable.transpose().dot(inputmovie['rating'])
userprofile
```

```python
genertable = moovierswithgeneres_df.set_index(moovierswithgeneres_df['movieId'])
genertable = genertable.drop('movieId',axis=1).drop('title',axis=1).drop('genres',axis=1).drop('year',axis=1)
genertable.head()
```

```python
genertable.shape
```

```python
recomendationaltable = ((genertable*userprofile).sum(axis=1))/(userprofile.sum())
recomendationaltable.head()
```

```python
recomendationaltable = recomendationaltable.sort_values(ascending=False)
recomendationaltable.head()
```
