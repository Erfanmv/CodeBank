```python
import pandas as pd
country = pd.read_csv("country.csv" , header = 1)
country =  country.rename(columns={'CountryName':"name",'Population growth':'growth','Total population':'pop','Area (sq. km)':'area'})
# country = country.set_index('name')
country.drop(['CountryCode' , '1'] ,  axis=1 , inplace=True)
country
country.rename(index=country.name , inplace=True)
# country = country.set_index('name')
# country.drop('name', axis=1 , inplace=True)
# country = country.set_index('name')
max_pop=country['pop'].max()
country['pop'][country['pop']==max_pop]
country.drop('World', axis=0 , inplace=True)
country['growth']=pd.to_numeric(country['growth'], errors='coerce')
country['pop']=pd.to_numeric(country['pop'], errors='coerce')
country['area']=pd.to_numeric(country['area'], errors='coerce')
# country = country.set_index('name')
country
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,10])
y = np.array([20,100])
font1 = {'family':'serif' , 'color' : 'blue', 'size':20}
font2 = {'family':'sans serif' , 'color' : 'darkred', 'size':15}
plt.grid()
#- -- : --. line style
#0+*X*.,sd
#rgpcmykw color
plt.plot(x,y , '*-.b' ,  ms =  15 , mec ='b' , mfc = 'y' , linewidth = 5)
plt.title('mft' , loc = 'left')
plt.xlabel('growth' ,  fontdict=font2)
plt.ylabel('pop' , fontdict=font2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
lesson = ['python', 'react', 'js', 'npdejs']
mft = [1000 , 700, 500, 200]
azad = [800 , 400 ,300 ,150]
plt.title('mft & azad')
# plt.margins(0.4)
# plt.figure(figsize=(12,6), dpi= 100)
plt.xlabel('lesson' )
plt.ylabel('score' )
plt.margins(0.4)
plt.plot(lesson,mft ,ls = '--', marker = '+', mew = 3 )
plt.plot(lesson,azad ,ls = '--', marker = '+', lw = 3 )
plt.grid()
plt.annotate('mftpython', xytext = (0.5,650), xy=(0,990), arrowprops={'facecolor':'red', 'width':4 },fontsize  =12 )
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(lesson,mft)
plt.title('mft')
plt.subplot(2,1,2)
plt.plot(lesson,azad)
plt.title('azad')
```
